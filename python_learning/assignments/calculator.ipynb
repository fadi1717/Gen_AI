```python
def add(num1: int, num2: int) -> int:
    return num1 + num2

def subtract(num1: int, num2: int) -> int:
    return num1 - num2

def multiply(num1: int, num2: int) -> int:
    return num1 * num2

def divide(num1: int, num2: int) -> float:
    if num2 == 0:
        return "Error: Division by zero is not allowed."
    else:
        return num1 / num2

def power(num1: int, num2: int) -> int:
    return num1 ** num2

def modulus(num1: int, num2: int) -> int:
    return num1 % num2

def get_operation_choice() -> str:
    print("\nSelect operation:")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Power (num1^num2)")
    print("6. Modulus")

    return input("Enter your choice (1-6): ")

def get_two_numbers() -> tuple[int, int]:
    num1 = int(input("Enter the first number (integer): "))
    num2 = int(input("Enter the second number (integer): "))
    return num1, num2

def calculator() -> None:
    while True:
        choice = get_operation_choice()

        if choice in ['1', '2', '3', '4', '5', '6']:
            num1, num2 = get_two_numbers()
            print(f'\nNumber1 is {num1}')
            print(f'Number2 is {num2}')
            

        if choice == '1':
            print(f"Result: {add(num1, num2)}")
            print('Selected Choice Was Addition')

        elif choice == '2':
            print(f"Result: {subtract(num1, num2)}")
            print('Selected Choice Was Subtract')

        elif choice == '3':
            print(f"Result: {multiply(num1, num2)}")
            print('Selected Choice Was Multiplication')

        elif choice == '4':
            result = divide(num1, num2)
            print(f"Result: {result}")
            print('Selected Choice Was Division')

        elif choice == '5':
            print(f"Result: {power(num1, num2)}")
            print('Selected Choice Was Power')

        elif choice == '6':
            print(f"Result: {modulus(num1, num2)}")
            print('Selected Choice Was Modulus')

        else:
            print("Invalid choice. Please select a valid operation.")

        another_calculation = input("Do you want to perform another calculation? (yes/no): ")
        if another_calculation.lower() != 'yes':
            break


calculator()
```
